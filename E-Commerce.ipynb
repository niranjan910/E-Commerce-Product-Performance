## Data Cleaning

```python
df.isnull().sum()
```

**Product Price**

```python
sns.histplot(data=df,x = "Product_Price",kde=True)
```

```python
df['Product_Price'].skew()
```

```python
df['Product_Price'] = df['Product_Price'].fillna(df['Product_Price'].median())
```

**Discount Rate**

```python
sns.histplot(data=df,x = "Discount_Rate",kde=True)
```

```python
df['Discount_Rate'].skew()
```

```python
# As the data set is having a long right tail there might be outliers in the Discount rate column so we will replace the missing values with median
df['Discount_Rate'] = df['Discount_Rate'].fillna(df['Discount_Rate'].median())
```

**Product Rating**

```python
sns.histplot(data=df,x = "Product_Rating",kde=True)
```

```python
df['Product_Rating'].skew()
```

```python
# Number of to deliver might be robust to ouliers so we will replace the msiing values with median 
df['Product_Rating'] = df['Product_Rating'].fillna(df['Product_Rating'].median())
```

**Number of Reviews**

```python
df["Number_of_Reviews"]
```

```python
sns.histplot(data=df,x = "Number_of_Reviews",kde=True)
```

```python
# Number of to deliver might be robust to ouliers so we will replace the msiing values with median 
df['Number_of_Reviews'] = df['Number_of_Reviews'].fillna(df['Number_of_Reviews'].median())
```

```python
df["Number_of_Reviews"] = df["Number_of_Reviews"].astype('int')
```

**Stock_Availability**

```python
df["Stock_Availability"]
```

```python
df["Stock_Availability"].isna().sum()
```

```python
# Since its a categorical column we will replace the value by mode
df['Stock_Availability'] = df['Stock_Availability'].fillna(df['Stock_Availability'].mode()[0])
```

```python
df['Stock_Availability'].replace({0: 'No', 1: 'Yes'}, inplace=True)
```

**Days_to_Deliver**

```python
df["Days_to_Deliver"]
```

```python
sns.histplot(data=df,x = "Days_to_Deliver",kde=True)
```

```python
df['Days_to_Deliver'].skew()
```

```python
# Days to deliver might be robust to ouliers so we will replace the msiing values with median 
df['Days_to_Deliver'] = df['Days_to_Deliver'].fillna(df['Days_to_Deliver'].median())
```

```python
df["Days_to_Deliver"] = df["Days_to_Deliver"].astype('int')
```

**Return_Rate**

```python
df["Return_Rate"]
```

```python
sns.histplot(data=df,x = "Return_Rate",kde=True)
```

```python
df['Return_Rate'].skew()
```

```python
# Return rate might be robust to ouliers so we will replace the msiing values with median 
df['Return_Rate'] = df['Return_Rate'].fillna(df['Return_Rate'].median())
```

**Category_ID**

```python
df["Category_ID"]
```

```python
# Since Category id is a Categorical column we will replace the values with mode 
df['Category_ID'] = df['Category_ID'].fillna(df['Category_ID'].mode()[0])
```

```python
df["Category_ID"] = df["Category_ID"].astype('int')
```

```python
# To check for the duplicate roes in the data set
df.duplicated().sum()
```
